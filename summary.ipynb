#Minima of Action
###The action of a general dynamical system $\dot{x}=f(x)$ with measurements $y$ can formulated as
###$$A_0 = (x-y)^T \frac{R_m}{2}(x-y)+(\dot{x}-f(x))^T \frac{R_f}{2}(\dot{x}-f(x))$$

###where we assume the noise in the measurement are guassian distributed. 

###The probability of estimated state $X=(x_1,x_2,\dots,x_D)$ conditioned on the measurements $Y=(y_1,y_2,\dots,y_L)$ is
###$$P(X|Y)\propto\exp(-A_0)$$

###Therefore, we are interested in looking for the minima of the action which can achieve maximum probability of estimated state.

## Explore the minima of action when different numbers of measurements are present.

###Lorenz 96 model with D=5 is used. Acurrate data is obtained by integrating the model, and measurements is generated by adding white gaussian noise with $\sigma=0.5$


```python
%pylab inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("Agg")
font = {'size'   : 18}
matplotlib.rc('font', **font)
data = np.loadtxt('matlab_D5_Rf_M1/twin_data.dat')
fig = plt.figure(figsize=(16,10))
for i in range(5):
    ax1 = fig.add_subplot(3,2,i+1)
    ax1.plot(data[:,i],'b-',linewidth=2.0,label='data')
    #ax1.plot(assim[:,i],'r--',linewidth=2.0,label='estimate')
    ax1.set_ylabel('x%d'%(i+1))
    #box = ax1.get_position()
ax1.set_xlabel('time(0.025)')
ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

---

###We choose our data assimilation window to be $0\le t \le 4$, and $dt=0.025$, which contains 160 timesteps. The minimization of $A_0$ is in $160\times5=800$ dimensional space. The final local minimizer $X$ will satisfy the optimality condition
###$$\nabla A_0(X) = 0, ~~~\nabla^2 A_0(X) \text{ is positive semidefinte}$$
###The minimizations are started with 100 different intial guess paths $X^{(0)}$, which are composed by 800 random numbers generated from uniform distribution $[-10,10]$. 

###Each minimization is done in simulated annealing way, we fixed $R_m = 1/\sigma^2=4$, and $R_f=0.02$, and introduce a control parameter $\beta$ that multiplies $R_f$
###$$A_0 = (x-y)^T \frac{R_m}{2}(x-y)+\beta(\dot{x}-f(x))^T \frac{R_f}{2}(\dot{x}-f(x))$$
###We start with $\beta=1$ and find the minima,  and then increase $\beta$ by a factor of two, choose the minimizer obtained from last step as new intial guess and do the minimization again. We trace all the 100 minima under different values of $\beta$.

---

###The calculations are carried out with different numbers of measurements
1. one measurement: only $x_1$ is measured
1. two measurements: $x_1$ and $x_3$ are measured
1. three measurements: $x_1$, $x_3$ and $x_5$ are measured

```python
%pylab inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use("Agg")
font = {'size'   : 18}
matplotlib.rc('font', **font)

def plot_min(M=2):
    D = 5
    fig = plt.figure(figsize=(12,9))

    ax1 = fig.add_subplot(111)
    for B in range(1,23):
        minima = np.loadtxt('matlab_D5_Rf_M%d/data/minima_D%d_B%d.dat'%(M,D,B))
        if M==3:
            tmp = 1
        else:
            tmp = len(minima)
        ax1.scatter(B*np.ones(tmp), minima)
    ax1.set_title('Lorenz96 Dim=%d #Meas=%d'%(D,M))
    ax1.set_ylabel('Minimized Action')
    ax1.set_xlabel('Log2beta')
    ax1.set_yscale('log')
    plt.savefig('Action_D%d_M%d_beta.png'%(D,M),bbox_inches='tight')
```

```python
plot_min(3)
```

###When three measurements are present, 100 intial guesses converge to one minima.

```python
plot_min(2)
```

###As $\beta$ increases, we can see the seperation of global minima and local minima. 
###Let us assume $z(x)$ be the measurement error term and $c_i(x)=\dot{x}_i-f_i(x)$ be the discredited form of the model. The action is
###$$A_0=z(x)+\beta \frac12 R_f \sum_i c_i(x)^2$$
###The hessian matrix of $A_0$ is
###$$ \nabla ^2 A_0 =  \nabla^2 z(x) + \sum_i \beta R_f c_i(x) \nabla c_i(x) + \beta R_f (\nabla c(x))^T \nabla c(x) $$
###where $(\nabla c(x))^T \nabla c(x)$ is a positive definte matrix, increasing $\beta$ has the effect of adding positive curvature to the hessian matrix.

```python
plot_min(1)
```
